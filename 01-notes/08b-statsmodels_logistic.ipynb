```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model = smf.logit('survived ~ sex + age + fare + deck',
                 data = titanic)
results = model.fit()
results.summary()
```
